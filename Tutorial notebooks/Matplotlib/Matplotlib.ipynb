## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Graph

```python
x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5]
y = [0.05,0.1,0.15,0.19,0.23,0.27,0.31,0.35,0.38,0.42]

#Resize your Graph (dpi means pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi=150)

#Line 1

#Keybword argument notation
#plt.plot(x,y, label='2x', linestyle= linewidth=2, color='red', marker='.', markersize=10, markerfacecolor='black', markeredgecolor='black')

#Shorthand method
#fmt = '[marker][line][color]'

plt.plot(x,y, 'b.-', label='2x')

#Line 2

#Select intervals we want to plot points at
x2 = np.arange(0,4.5,0.5)

#Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2' )

#Plot remainder of graph as dotted line
plt.plot(x2[5:], x2[5:]**2, 'r--')

#Add title(Specify font name and size with fontdict)
plt.title('1st Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize' : 20})

#X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#X and Y ticks(Scale of graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#Add a legend
plt.legend()

#Save figure (300dpi is good when saving so graph has high resolution)
plt.savefig('mygraph', dpi=300)

#Show plot
plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bar[0].set_hatch('/')
# bar[1].set_hatch('0')
# bar[2].set_hatch('*')


plt.savefig('barchart.png', dpi=300)

plt.show()
```

## REAL WORLD EXAMPLES

### Example 1

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,5))

countries_in_graph = ['Australia', 'USA', 'Canada', 'South Korea']

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
# plt.plot(gas.Year, gas.Canada, 'r.-')
# plt.plot(gas.Year, gas['South Korea'], 'g.-')
# plt.plot(gas.Year, gas.Australia, 'y.-')

for country in gas:
    if country in countries_in_graph:
        plt.plot(gas.Year, gas[country], marker = '.', label=country)

plt.title('Gas Prices over time', fontdict={'fontweight':'bold', 'fontsize':18})

plt.xlabel('Year')
plt.ylabel('Price(in $)')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

### 2nd Example

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

### Histogram

```python
bin = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bin, color='#abcdef')

plt.xticks(bin)

plt.title('Distribution of overall ratings in FIFA 18')
plt.ylabel('No. of players')
plt.xlabel('Overall rating')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of Players in FIFA 18')

plt.savefig('Weight pie chart.png', dpi = 300)

plt.show()
```

### Pie Chart #2

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Players in FIFA 18 (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.savefig('Weight pie chart 2.png', dpi = 300)

plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Club Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')

plt.save('Club')
    
plt.show()
```
