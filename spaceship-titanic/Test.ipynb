```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head(20)
```

```python
df = df.dropna(subset=['CryoSleep'])
df['CryoSleep'] = df['CryoSleep'].astype(int)
df.head()
```

```python
df[['CryoSleep', 'Transported']]
```

```python
df['HomePlanet'].isna().sum()
```

```python
df = df.dropna(subset=['HomePlanet'])
df[['CryoSleep', 'Transported', 'HomePlanet']]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False)
```

```python
encoded_array = encoder.fit_transform(df[['HomePlanet']])
```

```python
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['HomePlanet']))
```

```python
encoded_df.head()
```

```python
df['HomePlanet_Earth'] = encoded_df['HomePlanet_Earth']
df['HomePlanet_Europa'] = encoded_df['HomePlanet_Europa']
df['HomePlanet_Mars'] = encoded_df['HomePlanet_Mars']
```

```python
df.head()
```

```python
new_df = df.drop(columns=['HomePlanet', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Name'])
new_df.head()
```

```python
new_df = new_df.dropna(subset=['VIP'])
new_df['VIP'] = new_df['VIP'].astype(int)
new_df['Transported'] = new_df['Transported'].astype(int)
new_df.head()
```

```python
new_df = new_df.dropna(subset=['Age'])
new_df['Age'] = new_df['Age'].astype(int)
new_df.head(20)
```

```python
new_df = new_df.dropna(subset=['Cabin'])
```

Extracting deck and side from cabin(cabin/num/side)

```python
x = []
y = []
for f in new_df['Cabin']:
    parts=f.split('/')
    x.append(parts[0])
    y.append(parts[2])
```

```python
new_df['Deck']=x
new_df['Side']=y
```

```python
new_df.head()
```

```python
new_df = new_df.drop(columns =['Cabin'])
new_df.head()
```

```python
new_df = new_df.rename(columns={'Detination': 'Destination'})
new_df.head()
```

```python
new_df = new_df.drop(columns =['PassengerId'])
new_df.head()
```
